```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
anomaly_data = r'C:/Users/hamed/Desktop/SemiFinalists/Project/Anomaly.csv'
anomaly = pd.read_csv(anomaly_data)
anomaly.head(5)
```

```python
plt.rcParams['figure.figsize'] = (30,8)
anomaly.plot.bar(x='year', y='anomaly')
plt.xlabel("Year", fontsize=22)
plt.ylabel("Anomaly", fontsize=22)
plt.title("Anomaly = Average Global Temperature - Average Temperature for Reference Years (1870-1900)", fontsize=22)
plt.legend().set_visible(False)
```
